```python
import pandas as pd
import os


df = pd.DataFrame()

if os.path.getsize('NTVSaM.csv'):
    df = pd.read_csv('NTVSaM.csv')
```

```python
# 1. Liste os 10 primeiros itens do dataset

display(df.head(10))
```

```python
# 2. Liste os últimos 10 itens do dataset. 

display (df.tail(10))
```

```python
#3. Qual o filme com o maior imdb_score e qual o filme com o pior imdb_score? 
max = df[(df['imdb_score'] == df['imdb_score'].max())]

titleMax = max['title'].tolist()[0]
scoreMax = max['imdb_score'].tolist()[0]

min = df[(df['imdb_score'] == df['imdb_score'].min())]

titleMin = min['title'].tolist()[0]
scoreMin = min['imdb_score'].tolist()[0]

print ("Filme com o maior score:")
print ("Titulo: ", titleMax)
print ("Score:", scoreMax)
print("------------------------")
print ("Filme com o menor score:")
print ("Titulo: ", titleMin)
print ("Score: ", scoreMin)
```

```python
#4. Qual a média de duração dos shows de TV e duração de filmes

onlyMovies = df.loc[df['type'] == 'MOVIE']
onlyShows = df.loc[df['type'] == 'SHOW']

meanMovies = float (onlyMovies.loc[:, 'runtime'].mean())
meanShows = float (onlyShows.loc[:, 'runtime'].mean())

print (f"Média de duração dos filmes: {meanMovies:.1f} minutos")
print (f"Média de duração dos shows: {meanShows:.1f} minutos")
```

```python
# 5. Qual o show de TV com maior duração por episódio e o filme com duração mais longa?

maxRuntimeMovie = onlyMovies.loc[onlyMovies['runtime'] == onlyMovies['runtime'].max()]

runtimeMaxRuntimeMovie = maxRuntimeMovie['runtime'].tolist()[0]
titleMaxRuntimeMovie = maxRuntimeMovie['title'].tolist()[0]

maxRuntimeShow = onlyShows.loc[onlyShows['runtime'] == onlyShows['runtime'].max()]

runtimeMaxRuntimeShow = maxRuntimeShow['runtime'].tolist()[0]
titleMaxRuntimeShow = maxRuntimeShow['title'].tolist()[0]

print ("Filme com a maior duaração:")
print ("Titulo: ", titleMaxRuntimeMovie)
print ("Duração:", runtimeMaxRuntimeMovie)
print("------------------------")
print ("Show com a maior duração média por episódio:")
print ("Titulo: ", titleMaxRuntimeShow)
print ("Score: ", runtimeMaxRuntimeShow)
```

```python
# 6. Qual ano apresentou a maior quantidade de filmes avaliados com imdb?

years = df['release_year']

display (years.mode().tolist()[0])
```

```python
# 7. Quais as categorias encontradas de age_certification e qual a quantidade de filmes para cada categoria?

certifications = df['age_certification'].unique().tolist()

for c in certifications:
    if str(c) != 'nan':
        titlesByCertification = df.loc[df['age_certification'] == c]
        print ("Categoria: ", c, " | Quantidade de filmes: ", len(titlesByCertification))
```

```python
# 8. Quais colunas apresentam campos vazios? Qual a quantidade de campos vazios?

columns = df[['id', 'title', 'type', 'description', 'release_year', 'age_certification', 'runtime', 'imdb_id', 'imdb_score', 'imdb_votes']].isnull().sum()
display(columns)

```

```python
# 9. Gere um gráfico que você considere interessante para representar a média de scores ao longo dos anos para filme e outro para shows.

import matplotlib.pyplot as plt

average_movie = onlyMovies.groupby('release_year')['imdb_score'].mean()
average_show = onlyShows.groupby('release_year')['imdb_score'].mean()

plt.figure(figsize=(10, 6))

# Plotar os filmes
plt.scatter(average_movie.index, average_movie.values, label='Filmes', color='blue', s=50)

# Adicionar título, rótulos e legenda
plt.title('Média das Notas IMDB ao Longo dos Anos')
plt.xlabel('Anos')
plt.ylabel('Média da Nota IMDB')
plt.legend()

# Mostrar o gráfico

plt.figure(figsize=(10, 6))
plt.scatter(average_show.index, average_show.values, label='Séries', color='green', s=50)

plt.title('Média das Notas IMDB ao Longo dos Anos')
plt.xlabel('Anos')
plt.ylabel('Média da Nota IMDB')
plt.legend()
```

```python
# 10. Gere um gráfico que você considere interessante para apontar a proporção de votantes para shows e filmes.

import matplotlib.pyplot as plt

movies_year = onlyMovies.groupby('release_year').size().sort_index()
shows_year = onlyShows.groupby('release_year').size().sort_index()

all_years = movies_year.index.union(shows_year.index)

movies_year = movies_year.reindex(all_years, fill_value=0)
shows_year = shows_year.reindex(all_years, fill_value=0)

plt.plot(movies_year.index, movies_year.values, color='blue', marker='o', label='Filmes')
plt.plot(shows_year.index, shows_year.values, color='green', marker='o', label='Shows')
plt.title('Número de Filmes e Shows Lançados por Ano')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Quantidade')
plt.legend()

plt.show()



```
